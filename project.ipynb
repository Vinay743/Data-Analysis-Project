# Import Libraries


```python
import numpy as np
import pandas as pd
```

# Load the Data

```python
df = pd.read_csv("ICT_Subdimension_Dataset new.csv")
df.head()
df.info()
df

```

# Clean Dataset

```python
# Check missing value
df.isnull().sum()
# Drop rows with too many missing values
df = df.dropna()
# Remove duplicates
df = df.drop_duplicates()
#Info
df.info()
```

# Basic Statistics

```python
df.describe()
```

# Numpy Operations

```python
#Select numeric column
n = df.select_dtypes(include=['int64','float64']).columns
n
```

```python
#Compute z-score 
for col in n:
    df[col+'_zscore'] = (df[col] - np.mean(df[col]))/np.std(df[col])
df.head()
```

Top 10 cities by "Household Internet Access(%)"

```python
c = df.sort_values(by="Household Internet Access (%)",ascending=False).head(10)
c
```

# Import Data Visualization Libraries

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

Correlation Heatmap

```python

plt.figure(figsize=(12,8))
sns.heatmap(df[n].corr(),annot=True,fmt=".2f")
plt.title("ICT Metrics Correlation Heatmap")
plt.show()
```

Top 10 Cities by Household Internet Access in 2024


```python
# Filter latest year
latest_year = df["Year"].max()
df_latest = df[df["Year"] == latest_year]

# Get top 10 cities for the latest year
top10 = df_latest.sort_values(by="Household Internet Access (%)", ascending=False).head(10)

# Pie chart
plt.figure(figsize=(7,7))
plt.pie(
    top10["Household Internet Access (%)"],
    labels=top10["City"],
    autopct="%0.1f%%",
    shadow=True
)
plt.title(f"Top 10 Cities by Household Internet Access in {latest_year}")
plt.show()
```

Internet Access over Years

```python


years = sorted(df["Year"].unique())
plt.figure(figsize=(18, 12))

for i, year in enumerate(years, 1):
    df_year = df[df["Year"] == year]

    plt.subplot(2, 3, i)
    plt.bar(df_year["City"], df_year["Household Internet Access (%)"])
    plt.title(f"Household Internet Access - {year}")
    plt.xticks(rotation=90)
    plt.ylabel("Internet Access (%)")
    plt.tight_layout()

plt.show()
```

# SQL Connection and Query Execution

```python
import sqlite3 as sql
```

```python
conn  = sql.connect("New ICT.db")
```

```python
df.to_sql("ict_metrices",conn,if_exists="replace",index=False)
```

```python

query1 = """
SELECT City, [Wireless Broadband Coverage 4G (%)]
FROM ict_metrices
ORDER BY [Wireless Broadband Coverage 4G (%)] DESC
LIMIT 5;
"""

pd.read_sql_query(query1, conn)
```

```python
query2= """
SELECT Year,AVG([Smart Water Meters (%)]) AS avg_water_meters
FROM ict_metrices
GROUP BY Year;
"""
pd.read_sql_query(query2,conn)
```

```python
query3="""
SELECT City,[e-Government (%)]
FROM ict_metrices
WHERE [e-Government (%)]>40
ORDER BY [e-Government (%)] DESC;
"""
pd.read_sql_query(query3,conn)
```
